```python
import tabula as tb
import pandas as pd
import numpy as np
```

```python
dfs_complete = tb.read_pdf("pdf/2016_EurA1c_report.pdf", pages='5', multiple_tables=True)
df = dfs_complete[1].iloc[2:].reset_index(drop=True) #hoeft maar 1x uitgevoerd te worden
df
```

```python
df.columns = ["Manufacturer", "Waarde1", "Waarde2temp", "Waarde2", "Waarde3"]
df
```

```python
temptonormaldf = df.copy()

for index, row in temptonormaldf.iterrows():
    if row["Waarde2temp"] != np.nan:
        temptonormaldf.loc[index, "Waarde2"] = row["Waarde2temp"] + " " + row["Waarde2"]

temptonormaldf.drop("Waarde2temp", axis=1, inplace=True)

temptonormaldf
```

```python
# # Dubbele titel rijen opschonen
# newdf = df.copy()

# for index, row in newdf.iterrows():
#     # print(row['EurA1c 2020-1 36.5 (35.9-37.1) mmol/mol n Mean Bias CV%'])
#     if(row['Waarde1'] is np.nan and newdf.at[index+2, 'Waarde1'] is np.nan):
#         newdf.at[index, 'Waarde1'] = newdf.at[index+1, 'Waarde1']
#         newdf.at[index, 'Waarde2'] = newdf.at[index+1, 'Waarde2']
#         newdf.at[index, 'Waarde3'] = newdf.at[index+1, 'Waarde3']
#         newdf.at[index, 'Manufacturer'] = newdf.at[index, 'Manufacturer'] + ' ' + newdf.at[index+2, 'Manufacturer']

# newdf = newdf.dropna().reset_index(drop=True)

# Dubbele titel rijen opschonen
newdf = temptonormaldf.copy()

for index, row in newdf.iterrows():
    if(row['Waarde1'] is np.nan and (index+1) < len(newdf)):
        newdf.at[index, 'Waarde1'] = newdf.at[index+1, 'Waarde1']
        newdf.at[index, 'Waarde2'] = newdf.at[index+1, 'Waarde2']
        newdf.at[index, 'Waarde3'] = newdf.at[index+1, 'Waarde3']
        newdf.at[index, 'Manufacturer'] = newdf.at[index, 'Manufacturer'] + ' ' + newdf.at[index+1, 'Manufacturer']

        newdf.at[index+1, 'Waarde1'] = np.nan

newdf = newdf.dropna().reset_index(drop=True)

```

```python
newdf
```

```python
cleaneddf = newdf.copy()

for index, row in cleaneddf.iterrows():
    
    for x in range(1, 4):
        waarde = row['Waarde' + str(x)].split()
        
        for y, waardeEenheid in enumerate(waarde):
            columns = [
                ['1_n', '1_Mean', '1_Bias', '1_CV'],
                ['2_n', '2_Mean', '2_Bias', '2_CV'],
                ['Mean_Bias', 'Mean_CV']
            ]

            cleaneddf.at[index, columns[x-1][y]] = waardeEenheid

cleaneddf = cleaneddf.drop(columns=["Waarde1", "Waarde2", "Waarde3"])
cleaneddf = cleaneddf[['Manufacturer', '1_n', '1_Mean', '1_Bias', '1_CV', '2_n', '2_Mean', '2_Bias', '2_CV', 'Mean_Bias', 'Mean_CV']]
cleaneddf
```

```python
cleaneddf.to_excel("data/2017_Frozen_EurAAA1c_report.xlsx", index=False)
```
